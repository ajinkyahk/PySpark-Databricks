#pyspark-write-hive-table

```
columns = ["id", "name","age","gender"]

# Create DataFrame 
data = [(1, "James",30,"M"), (2, "Ann",42,"F"),
    (3, "Jeff",41,"M"),(4, "Jennifer",20,"F")]

sampleDF = sc.parallelize(data).toDF(columns)
```

```
# Create Hive internal table
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType
df = sampleDF.select(col("id").cast('int'), col("name"), col("age").cast("int"), col("gender"))
df.printSchema()
df.write.mode("overwrite").saveAsTable("ct.sampleTable")
```
