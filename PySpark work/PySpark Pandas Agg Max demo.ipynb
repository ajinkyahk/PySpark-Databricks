```
data = [(1, 'male', 'ajinkya', 'Pune', 24000), 
        (2, 'male', 'aneesh', 'Ichalkaranji', 38000),
        (3, 'female', 'pooja', 'Delhi', 34000),
        (4, 'male', 'suresh', 'Panji', 20000),
        (5, 'female', 'riya', 'Kolhapur', '12000')
       ]
columns = ['id', 'gender', 'name', 'city', 'salary']

df = spark.createDataFrame(data=data, schema=columns)

df.printSchema()
```

```
df.show()
```

```
from pyspark.sql.functions import *
aggdf = df.groupBy('gender').agg(max('city'))

aggdf.show()
```
