```
with open('/cofig_demo/config.json', 'r') as json_file:
    dict1 = json.load(json_file)
print(dict1.get('phone_no'))
```

```
df = spark.read.csv('dbfs:/FileStore/cofig_demo/emp_data.csv', header=True)
```

```
display(df)
```

```
# check mob no validation 
df = df.select('phone_no').toPandas()
phone_nos = df['phone_no'].to_list()
pattern = dict1.get('phone_no')
result = [True if re.match(pattern, x)!= None else False for x in phone_nos ]
print(result)
```

```
if np.all(result)==True:
    print(f'Phone numbers are valid')
else:
    print(f'Phone numbers are not valid')
```
