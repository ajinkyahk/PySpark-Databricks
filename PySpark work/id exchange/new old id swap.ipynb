```
sample1DF = spark.read.format("com.crealytics.spark.excel").option("header", "true").option("inferSchema", "true").option("treatEmptyValuesAsNulls", "false").option("dataAddress", "'Sheet1'!A1:C6").load("dbfs:/FileStore/mysheet/new_old_id_demo.xlsx")

display(sample1DF)
```

```
sample2DF = spark.read.format("com.crealytics.spark.excel").option("header", "true").option("inferSchema", "true").option("treatEmptyValuesAsNulls", "false").option("dataAddress", "'Sheet1'!F1:G6").load("dbfs:/FileStore/mysheet/new_old_id_demo.xlsx")

display(sample2DF)
```

```
from pyspark.sql.functions import *
```

```
sample1DF = sample1DF.join(sample2DF, on=sample1DF.id==sample2DF.oldid, how='left')
display(sample1DF)
```

```
sample1DF = sample1DF.withColumn('ids', when(sample1DF.newid.isNull() == True, sample1DF.id).otherwise(sample1DF.newid))
display(sample1DF)
```

```
sample1DF = sample1DF.drop('id', 'newid', 'oldid').withColumnRenamed('ids', 'id')
sample1DF = sample1DF.select('id', 'name', 'city')
display(sample1DF)
```
